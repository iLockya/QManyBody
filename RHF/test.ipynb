```julia
include("RHFSCF.jl")
include("gaussian.jl")
using LinearAlgebra
using TensorOperations
using Plots
```

```julia
# check H atom.
E₀ = ∫(STO3G(),STO3G();type = "kinetic") + ∫(STO3G(),STO3G();type = "nucleus")
```

```julia
dsr = [0.7,1.0,1.3,1.4,1.5,2.0,3.0,4.0]
BEr = []
for d in dsr
    system = init_H2(d)
    RHFSCF!(system)
    push!(BEr,system.E- 2E₀)
end
ds = range(0.5,5,100)
BE = []
for d in ds
    system = init_H2(d)
    RHFSCF!(system)
    push!(BE,system.E- 2E₀)
end
```

```julia
theme(:wong)
plot(ds,BE,linewidth=2,label="Binding Curve",xlabel="d (a.u.)",ylabel="Binding Energy (a.u.)",background_color_outside=false)
scatter!(dsr,BEr,label="Homework points")
title!("Energy Curve for STO-3G H₂")
# savefig("STO-3G_H2.png")
```
